## Análise de Dados dos Deputados

-  Quantos deputados por sexo?
-  Quantos deputados por estado?
-  Qual partido possui mais deputados em São Paulo?
-  Qual partido possui mais deputados na Bahia?
-  Qual estado possui o maior número de deputadas?
-  Qual partido consta com mais mulheres como deputadas?

```
#spark
spark
```

```
#Carregando dados
df = spark.read.csv('/FileStore/tables/deputados-4.csv',header = True,encoding = 'Latin1',sep=';')
```

```
df.show()
```

```
df.printSchema()
```

```
df.count()
```

### Quantos deputados por sexo?

```
#deputados por sexo
df.groupBy(df.sexo).count().orderBy('count').show()
```

### Quantos deputados por estado?

```
#deputados por estado 
df.groupBy(df.uf).count().orderBy('count').orderBy('count', ascending = False).show()
```

### Qual partido possui mais deputados em São Paulo?

```
# partido por estado em SP
df.where(df.uf == 'SP').groupBy(df.partido).count().orderBy('count').orderBy('count', ascending = False).show()
```

### Qual partido possui mais deputados na Bahia?

```
# partido na BA
df.where(df.uf == 'BA').groupBy(df.partido).count().orderBy('count').orderBy('count', ascending = False).show()
```

### Qual estado possui o maior número de deputadas?

```
# estado com maior número de deputadas
df.where(df.sexo == 'feminino').groupBy(df.uf).count().orderBy('count').orderBy('count', ascending = False).show()
```

### Qual partido consta com mais mulheres como deputadas?

```
# partido com maior número de deputadas
df.where(df.sexo == 'feminino').groupBy(df.partido).count().orderBy('count').orderBy('count', ascending = False).show()
```
